```python
def New_M_Matrix(dimension,bandwidth):
    A = np.eye(dimension)
    for i in range(0, -bandwidth, -1):
        A += np.eye(dimension, k = i -1)
    for i in range(dimension):
        temp = np.sum(A[i])
        for k in range(dimension):
            if A[i][k] == 1:
                A[i][k] = 1/temp
    return A
```
